```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Posiciones de la pelota en el campo visual
positions = np.arange(1, 6)

# Activaciones por condición clínica
activaciones = {
    "Normal": [1, 2, 3, 2, 1],
    "TEA": [1, 0, 2, 0, 1],
    "Daño occipital": [0, 0, 0, 0, 0],
    "Agnosia visual": [1, 2, 3, 2, 1],
    "Hemianopsia": [0, 0, 3, 2, 1],
    "Envejecimiento": [0.8, 1.5, 2.5, 1.5, 0.8]
}

# Colores personalizados
colores = {
    "Normal": "black",
    "TEA": "blue",
    "Daño occipital": "red",
    "Agnosia visual": "purple",
    "Hemianopsia": "green",
    "Envejecimiento": "gray"
}

# Función para mostrar el gráfico según la condición seleccionada
def mostrar_grafico(condicion):
    activacion = activaciones[condicion]
    plt.figure(figsize=(8,5))
    plt.plot(positions, activacion, marker='o', color=colores[condicion], label=condicion)
    plt.xlabel('Posición de la pelota en el campo visual')
    plt.ylabel('Activación de célula compleja V1')
    plt.title(f'Activación en V1 - Condición: {condicion}')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Menú desplegable interactivo
menu = widgets.Dropdown(
    options=list(activaciones.keys()),
    description='Condición clínica:',
    style={'description_width': 'initial'}
)

# Activar interactividad
widgets.interact(mostrar_grafico, condicion=menu)
```

📊 ¿Qué representa el eje X: “posición de la pelota”?
El eje X simula la trayectoria espacial de la pelota a lo largo del campo visual. Cada punto (1 a 5) representa una posición secuencial en la que la pelota se encuentra mientras se desplaza, por ejemplo, de izquierda a derecha.

Posición 1: pelota en el extremo izquierdo del campo visual.

Posición 3: pelota en el centro.

Posición 5: pelota en el extremo derecho.

Este eje permite modelar cómo se activan las células complejas de V1 a medida que la pelota se mueve por el campo visual.
