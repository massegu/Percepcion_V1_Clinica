```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Simulación de activación de células simples
def simulate(condition):
    positions = np.arange(1, 6)

    if condition == "Normal":
        activation = [1, 2, 3, 2, 1]
    elif condition == "TDAH":
        activation = [0.5, 1, 1.5, 1, 0.5]
    elif condition == "TEA":
        activation = [1, 0, 2, 0, 1]
    elif condition == "Daño occipital":
        activation = [0, 0, 0, 0, 0]
    elif condition == "Agnosia visual":
        activation = [1, 2, 3, 2, 1]  # activación intacta pero sin reconocimiento
    elif condition == "Hemianopsia":
        activation = [0, 0, 3, 2, 1]  # pérdida en hemicampo izquierdo
    elif condition == "Envejecimiento":
        activation = [0.8, 1.5, 2.5, 1.5, 0.8]

    plt.figure(figsize=(6,4))
    plt.plot(positions, activation, marker='o', label=f'Condición: {condition}')
    plt.xlabel('Posición de la pelota')
    plt.ylabel('Activación de célula compleja')
    plt.title('Percepción del movimiento en V1')
    plt.grid(True)
    plt.legend()
    plt.show()

# Widget interactivo
dropdown = widgets.Dropdown(
    options=["Normal", "TDAH", "TEA", "Daño occipital", "Agnosia visual", "Hemianopsia", "Envejecimiento"],
    description='Condición:',
    style={'description_width': 'initial'}
)

widgets.interact(simulate, condition=dropdown)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Posiciones de la pelota
positions = np.arange(1, 6)

# Activaciones por condición
activaciones = {
    "Normal": [1, 2, 3, 2, 1],
    "TEA": [1, 0, 2, 0, 1],
    "Daño occipital": [0, 0, 0, 0, 0],
    "Agnosia visual": [1, 2, 3, 2, 1],
    "Hemianopsia": [0, 0, 3, 2, 1],
    "Envejecimiento": [0.8, 1.5, 2.5, 1.5, 0.8]
}

# Colores personalizados
colores = {
    "Normal": "black",
    "TEA": "blue",
    "Daño occipital": "red",
    "Agnosia visual": "purple",
    "Hemianopsia": "green",
    "Envejecimiento": "gray"
}

# Gráfico superpuesto
plt.figure(figsize=(10,6))
for condicion, activacion in activaciones.items():
    plt.plot(positions, activacion, marker='o', label=condicion, color=colores[condicion])

plt.xlabel('Posición de la pelota')
plt.ylabel('Activación de célula compleja V1')
plt.title('Comparativa de activación en V1 según condición clínica')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

📊 ¿Qué representa el eje X: “posición de la pelota”?
El eje X simula la trayectoria espacial de la pelota a lo largo del campo visual. Cada punto (1 a 5) representa una posición secuencial en la que la pelota se encuentra mientras se desplaza, por ejemplo, de izquierda a derecha.

Posición 1: pelota en el extremo izquierdo del campo visual.

Posición 3: pelota en el centro.

Posición 5: pelota en el extremo derecho.

Este eje permite modelar cómo se activan las células complejas de V1 a medida que la pelota se mueve por el campo visual.
